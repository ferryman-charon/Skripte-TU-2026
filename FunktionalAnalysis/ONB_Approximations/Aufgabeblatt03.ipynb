## Legendre Plynomials

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
P0 = lambda x: 1
P1 = lambda x: x
P2 = lambda x: (3*x**2 - 1)
P3 = lambda x: (5*x**3 - 3*x)
P4 = lambda x: (63*x**5 - 70*x**3 + 15*x)  # Grad 5 (falls gewnscht)

#
coeffs = [0.5, 0.75, 0, -7/32, 11/(16*8)] 

x = np.linspace(-1, 1, 1000)

# Berechne sukzessive Partialsummen
y_partial = np.zeros_like(x)

plt.figure(figsize=(8,5))

y_partial += coeffs[0]*P0(x)
plt.plot(x, y_partial, label='n=0', linestyle='--')

y_partial += coeffs[1]*P1(x)
plt.plot(x, y_partial, label='n=1', linestyle='--')

y_partial += coeffs[3]*P3(x)
plt.plot(x, y_partial, label='n=3', linestyle='--')

y_partial += coeffs[4]*P4(x)
plt.plot(x, y_partial, label='n=4', linestyle='-')

plt.plot(x, np.where(x>0, 1,0), color='black', linewidth=2)
plt.grid(True, linestyle='--')
plt.legend()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Sukzessive Annherung von f(x) durch Legendre-Partialsummen")
plt.show()
```

## ONS aproximations

```python
x = np.linspace(-np.pi, np.pi, 1000)

y_partial = 6*np.sin(x) - 3*np.sin(2*x) + 2*np.sin(3*x)

plt.figure(figsize=(8,5))

plt.plot(x, y_partial, label='Approximation', linestyle='-')

plt.plot(x, 3*x, color='black', linewidth=2, label='Function')
plt.grid(True, linestyle='--')
plt.legend()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Sukzessive Annherung von f(x) durch ein ONS")
```

```python
x = np.linspace(-1, 1, 1000)

y_partial = 3/np.e * x

plt.figure(figsize=(8,5))

plt.plot(x, y_partial, label='Approximation ONS', linestyle='-')
plt.plot(x, x, label='Approximation Taylor', linestyle='--')

plt.plot(x, np.sinh(x), color='black', linewidth=2, label='Function')
plt.grid(True, linestyle='--')
plt.legend()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Sukzessive Annherung von f(x) durch ein ONS")
```

```python
x = np.linspace(-1, 1, 1000)

y_partial = 3/np.e * x

plt.figure(figsize=(8,5))

plt.plot(x, abs(np.sinh(x) - y_partial), label='Approximation ONS', linestyle='-')
plt.plot(x, abs(np.sinh(x) - x), label='Approximation Taylor', linestyle='--')

plt.grid(True, linestyle='--')
plt.legend()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Differenz der Aproximation zu f(x)")
```

## Cauchy Folge Proof

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1, 1, 1000)
ns = np.array([1,2,3,4,10])

def fn(x, n):
    return abs(x)**(1+1/n)

for n in ns:
    plt.plot(x, fn(x, n))

plt.xlabel("x")
plt.ylabel("y")
```
